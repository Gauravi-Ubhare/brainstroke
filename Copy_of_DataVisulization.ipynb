```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#What does the histogram reveal about the distribution of BMI in the dataset?
sns.histplot(data=df, x='bmi', bins=6, kde=True, color='green',edgecolor='blue')
plt.xlabel('Bmi')
plt.ylabel('Frequency')
plt.show()
```

```python
#What is the proportion of individuals who are ever married compared to those who are not?
marriedornot=df['ever_married'].value_counts()
marriedornot.plot(kind='pie',autopct='%1.1f%%')
plt.title('Ever Married or Not')
plt.show()
```

```python
#What is the distribution of ages among individuals in the dataset?
plt.figure(figsize=(6,7))
sns.histplot(data=df, x='age',bins=20, kde=True, color='red',edgecolor='blue')
plt.title('Frequency of age')
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.show()
```

```python
#What is the distribution of individuals across different residence types in the dataset?
sns.countplot(x='Residence_type', data=df)
plt.title('Count of Residence Types')
plt.xlabel('Residence Type')
plt.ylabel('Count')
plt.show()
```

```python
#What does the boxplot of BMI reveal about the distribution of BMI values in the dataset?
df.boxplot(column=['bmi'])
plt.title('Boxplot of BMI')
plt.ylabel('BMI Value')
plt.show()

```

```python
#What is the distribution of individuals among different smoking statuses in the dataset?
plt.figure(figsize=(6,6))
sns.countplot(x='smoking_status',data=df,palette='rocket_r',order=df['smoking_status'].value_counts().index)
plt.xticks(rotation=0)
plt.title('CountPlot')
plt.show()
```

```python
#How are the distributions of 'avg_glucose_level' and 'age' represented in the boxplots?
#BOXPLOT
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 1)
sns.boxplot(y='avg_glucose_level', data=df)
plt.title('Boxplot of avg_glucose_level')
plt.subplot(2, 2, 2)
sns.boxplot(y='age', data=df)
plt.title('Boxplot of age')
plt.tight_layout()
plt.show()
```

```python
#How does age relate to smoking status among individuals in the dataset?
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x='smoking_status',y='age',color='orange')
plt.title("Smoking satuts ")
plt.xlabel("Smoking_status")
plt.ylabel("Age")
plt.show()
```

```python
#How does the average glucose level vary between male and female individuals in urban and rural areas ('Residence_type')?
br=df['Residence_type'].sort_values(ascending=True)
plt.figure(figsize=(6,6))
sns.barplot(x=br, y='avg_glucose_level', hue='gender',data=df,errorbar=None, palette='Set1')
plt.xlabel('Residence type')
plt.ylabel('Avg_glucose level')
plt.title("Avg_glucose level in Male and Female")
plt.legend(loc='upper center',ncol=2)
plt.show()
```

```python
#Which worker types has lowest worker(patients)?
data = {'work_type': ['Private', 'Self_employed', 'Children', 'Govt_job'],
        'count': [2860,804,673,644]}
df = pd.DataFrame(data)
e=[0.0,0.0,0.0,0.3]
colors = ['coral', 'blue', 'green', 'skyblue']
plt.figure(figsize=(6, 6))
plt.pie(df['count'], labels=df['work_type'], colors=colors, autopct='%1.1f%%', startangle=140, explode=e)
plt.title('Worker types Distribution')
plt.axis('equal')
plt.show()
```

```python
#How does the average glucose level vary among different smoking statuses?
sns.lineplot(x='smoking_status', y='avg_glucose_level', label='Smoking Status', data=df)
plt.legend()
plt.xlabel('Smoking Status')
plt.ylabel('Average Glucose Level')
plt.title('Average Glucose Level by Smoking Status')
plt.xticks(rotation=45)
plt.show()
```

```python
#What are the correlations between the variables 'stroke,' 'bmi' (body mass index), 'heart_disease,' and 'hypertension'?
corr_matrix=df[['stroke','heart_disease','hypertension']]
corr_data=corr_matrix.corr()
plt.figure(figsize=(6,6),dpi=100)
sns.heatmap(corr_data, annot=True, cmap='flare',fmt=".2f")
plt.title("correlation matrix heatmap")
plt.show()
```

```python
#How does BMI vary across different work types, and how does gender impact this variation?
plt.figure(figsize=(6,5))
sns.violinplot(data=df, x='work_type',y='bmi',hue='gender',palette="husl")
plt.xlabel('Wrok type')
plt.ylabel('Bmi')
plt.xticks(rotation=0)
plt.legend(loc='upper left',ncol=2)
plt.show()
```

```python
#What relationships or patterns can we observe between hypertension, heart disease, and stroke based on the pair plot?
sns.pairplot(df[['hypertension',	'heart_disease','stroke']], diag_kind='kde')
plt.suptitle("Pair Plot of Disease", y=1.01)
plt.show()
```
